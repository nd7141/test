```python
import pandas as pd

df = pd.read_csv('task2.txt', sep='\t', header=None)
df # here is original table
```

```python
N = len(df)
table = []
for j in range(len(df.columns)):
    column = []
    counts = df[j].value_counts()
    for num in df[j]:
        column.append(counts[num]/float(N))
    table.append(column)

df2 = pd.DataFrame(table)
df3 = df2.transpose()
df3 # here is the table with probabilities 
```

```python
df3.values.sort()
df3 # here is the table with sorted probabilities (from smallest to largest)
```
